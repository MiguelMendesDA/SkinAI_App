```python
import subprocess

def download_isic_archive():
    try:
        # Execute the command t
        subprocess.run(["isic", "image", "download",  "--search", 'benign_malignant:"benign"',"benign/"])
        subprocess.run(["isic", "metadata", "download"])
        print("ISIC archive downloaded successfully.")
    except Exception as e:
        print(f"Error downloading ISIC archive: {e}")

if __name__ == "__main__":
    download_isic_archive()
```
